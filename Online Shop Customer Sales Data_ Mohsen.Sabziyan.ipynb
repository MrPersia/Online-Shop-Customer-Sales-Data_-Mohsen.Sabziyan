```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten laden
file_path = r"D:\Weiterbildung - Data Analyst\Fabian Dozent\Woche 15 (01.12.2023)\Online Shop Customer Sales Data.csv"
data = pd.read_csv(file_path)

# 1. Durchschnittsalter der Kunden
average_age = data['Age'].mean()
print(f"Durchschnittsalter der Kunden: {average_age:.2f} Jahre")

plt.figure(figsize=(8, 6))  # Größe des Diagramms festlegen
# Histogramm für die Verteilung des Alters der Kunden erstellen
data['Age'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')  
plt.title('Verteilung des Alters der Kunden', fontsize=16)  # Titel des Diagramms mit größerer Schriftgröße
plt.xlabel('Alter', fontsize=14)  # Beschriftung der x-Achse mit größerer Schriftgröße
plt.ylabel('Anzahl der Kunden', fontsize=14)  # Beschriftung der y-Achse mit größerer Schriftgröße

# Hinzufügen von Rasterlinien und Gitterlinien für bessere Lesbarkeit
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='black', alpha=0.7)
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='gray', alpha=0.5)

# Anmerkung für den Durchschnittswert des Alters
average_age = data['Age'].mean()
plt.axvline(average_age, color='red', linestyle='--', linewidth=2, label=f'Durchschnittsalter: {average_age:.2f} Jahre')
plt.legend()

plt.tight_layout()  # Einstellungen für ein passendes Layout
plt.show()  # Diagramm anzeigen


#1.1 Durchschnittliche Anzahl der Einkäufe nach Alter


# Gruppieren der Daten nach Alter und Berechnung des Durchschnitts der Einkäufe für jedes Alter
avg_purchases_by_age = data.groupby('Age')['N_Purchases'].mean().reset_index()

# Finden des Alters mit den meisten Einkäufen
max_purchases_age = avg_purchases_by_age.loc[avg_purchases_by_age['N_Purchases'].idxmax()]
min_purchases_age = avg_purchases_by_age.loc[avg_purchases_by_age['N_Purchases'].idxmin()]
mean_purchases = avg_purchases_by_age['N_Purchases'].mean()

# Visualisierung der durchschnittlichen Anzahl der Einkäufe pro Alter
plt.figure(figsize=(10, 6))
plt.bar(avg_purchases_by_age['Age'], avg_purchases_by_age['N_Purchases'], color='skyblue')
plt.xlabel('Alter')
plt.ylabel('Durchschnittliche Anzahl der Einkäufe')
plt.title('Durchschnittliche Anzahl der Einkäufe nach Alter')
plt.axvline(x=max_purchases_age['Age'], color='red', linestyle='--', label=f'Max Einkäufe: {max_purchases_age["Age"]} Jahre')
plt.axvline(x=min_purchases_age['Age'], color='green', linestyle='--', label=f'Min Einkäufe: {min_purchases_age["Age"]} Jahre')
plt.legend()
plt.tight_layout()
plt.show()

# Ausgabe von Max, Min und Mean
print(f"Max Einkäufe: {max_purchases_age['Age']} Jahre - Durchschnittliche Anzahl: {max_purchases_age['N_Purchases']:.2f}")
print(f"Min Einkäufe: {min_purchases_age['Age']} Jahre - Durchschnittliche Anzahl: {min_purchases_age['N_Purchases']:.2f}")
print(f"Durchschnittliche Anzahl der Einkäufe insgesamt: {mean_purchases:.2f}")



# 2. Gesamteinnahmen und Anzahl der Einkäufe
total_revenue = data['Revenue_Total'].sum()
total_purchases = data['N_Purchases'].sum()
print(f"Gesamteinnahmen: {total_revenue:.2f}")
print(f"Gesamtanzahl der Einkäufe: {total_purchases}")

plt.figure(figsize=(8, 4))
data[['Revenue_Total', 'N_Purchases']].sum().plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Gesamteinnahmen und Anzahl der Einkäufe')
plt.xlabel('Kategorie')
plt.ylabel('Summe')
plt.tight_layout()
plt.show()

# 3. Beliebteste Zahlungsmethode
# Ersetzen der Werte in der Spalte 'Pay_Method' entsprechend der neuen Zuordnung
data['Pay_Method'] = data['Pay_Method'].replace({
    0: 'Karte',
    1: 'PayPal',
    2: 'Digitale Geldbörsen',
    3: 'Andere'
})

# Zählen der Häufigkeit der geänderten Zahlungsmethoden
payment_method_counts = data['Pay_Method'].value_counts()

# Ausgabe der häufigsten Zahlungsmethoden
print("Häufigste Zahlungsmethoden:")
print(payment_method_counts)

# Erstellung des Balkendiagramms für die Anzahl der verschiedenen Zahlungsmethoden
plt.figure(figsize=(8, 6))
payment_method_counts.plot(kind='bar', color='skyblue')
plt.title('Häufigste Zahlungsmethoden')
plt.xlabel('Zahlungsmethode')
plt.ylabel('Anzahl')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()


# 4. Verteilung der Kunden nach Geschlecht

data['Gender'] = data['Gender'].replace({
    0: 'Male',
    1: 'Female',
})

gender_distribution = data['Gender'].value_counts()
print("Verteilung nach Geschlecht:")
print(gender_distribution)

plt.figure(figsize=(6, 6))
gender_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen'])
plt.title('Verteilung nach Geschlecht')
plt.axis('equal')
plt.tight_layout()
plt.show()

# 5. Häufigkeit von Newsletter-Abonnements

data['Newsletter'] = data['Newsletter'].replace({
    0: 'No',
    1: 'Yes',
})
newsletter_counts = data['Newsletter'].value_counts()
print("Häufigkeit von Newsletter-Abonnements:")
print(newsletter_counts)

plt.figure(figsize=(6, 4))
newsletter_counts.plot(kind='bar', color='skyblue')
plt.title('Häufigkeit von Newsletter-Abonnements')
plt.xlabel('Newsletter abonniert')
plt.ylabel('Anzahl')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 6. Verwendung von Gutscheinen/Rabattcodes

data['Voucher'] = data['Voucher'].replace({
    0: 'No',
    1: 'Yes',
})
voucher_counts = data['Voucher'].value_counts()
print("Verwendung von Gutscheinen/Rabattcodes:")
print(voucher_counts)

plt.figure(figsize=(6, 4))
voucher_counts.plot(kind='bar', color='lightgreen')
plt.title('Verwendung von Gutscheinen/Rabattcodes')
plt.xlabel('Gutschein verwendet')
plt.ylabel('Anzahl')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 7. Häufigster verwendeter Webbrowser
# Ersetzen der Werte in der Spalte 'Browser' entsprechend der neuen Zuordnung
data['Browser'] = data['Browser'].replace({
    0: 'Chrome',
    1: 'Safari',
    2: 'Edge',
    3: 'Andere'
})

# Zählen der Häufigkeit der geänderten Browser
browser_counts = data['Browser'].value_counts()

# Ausgabe des häufigsten verwendeten Webbrowsers
print("Häufigster verwendeter Webbrowser:")
print(browser_counts)

# Erstellung des Balkendiagramms für die Anzahl der verschiedenen Browser
plt.figure(figsize=(8, 6))
browser_counts.plot(kind='bar', color='skyblue')
plt.title('Häufigster verwendeter Webbrowser')
plt.xlabel('Webbrowser')
plt.ylabel('Anzahl')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()


# 8. Durchschnittliche Zeit, die Kunden auf der Website verbringen
average_time_spent_seconds = data['Time_Spent'].mean()
average_time_spent_minutes = average_time_spent_seconds / 60  # Umrechnung von Sekunden in Minuten
print(f"Durchschnittliche Zeit auf der Website verbracht: {average_time_spent_minutes:.2f} Minuten")

max_time_spent_seconds = data['Time_Spent'].max()
max_time_spent_minutes = max_time_spent_seconds / 60  # Umrechnung von Sekunden in Minuten
print(f"Maximale Zeit auf der Website verbracht: {max_time_spent_minutes:.2f} Minuten")

min_time_spent_seconds = data['Time_Spent'].min()
min_time_spent_minutes = min_time_spent_seconds / 60  # Umrechnung von Sekunden in Minuten
print(f"Minimale Zeit auf der Website verbracht: {min_time_spent_minutes:.2f} Minuten")


# 9. Verteilung der Einkäufe nach Datum
data['Purchase_DATE'] = pd.to_datetime(data['Purchase_DATE'])

# Verteilung der Einkäufe nach Monat visualisieren
plt.figure(figsize=(8, 6))
monthly_purchases = data['Purchase_DATE'].dt.month.value_counts().sort_index()
monthly_purchases.plot(kind='bar', color='skyblue')
plt.title('Verteilung der Einkäufe nach Monat')
plt.xlabel('Monat')
plt.ylabel('Anzahl der Einkäufe')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Ermitteln des Monats mit den meisten Einkäufen
most_purchased_month = monthly_purchases.idxmax()  # Index des Monats mit den meisten Einkäufen finden
print(f"Der Monat mit den meisten Einkäufen ist: {most_purchased_month}")

# 9.1 

# Verteilung der Einkäufe nach Tag visualisieren
plt.figure(figsize=(10, 6))
daily_purchases = data['Purchase_DATE'].dt.day.value_counts().sort_index()
daily_purchases.plot(kind='bar', color='skyblue')
plt.title('Verteilung der Einkäufe nach Tag')
plt.xlabel('Tag im Monat')
plt.ylabel('Anzahl der Einkäufe')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Ermitteln des Tags mit den meisten Einkäufen
most_purchased_day = daily_purchases.idxmax()  # Index des Tages mit den meisten Einkäufen finden
print(f"Der Tag im Monat mit den meisten Einkäufen ist: {most_purchased_day}")


# 10. Verteilung der Anzahl der Einkäufe nach Wochentagen

data['Purchase_DATE'] = pd.to_datetime(data['Purchase_DATE'])

# Extrahieren von Wochennummern, Tagen und Wochentagen in neue Spalten
data['Week_Number'] = data['Purchase_DATE'].dt.isocalendar().week
data['Day_of_Week'] = data['Purchase_DATE'].dt.day
data['Weekday_Name'] = data['Purchase_DATE'].dt.day_name()

print(data['Weekday_Name'].describe())

# Diagramm für die Verteilung der Anzahl der Einkäufe nach Wochentagen
plt.figure(figsize=(8, 6))
data['Weekday_Name'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Verteilung der Anzahl der Einkäufe nach Wochentagen')
plt.xlabel('Wochentag')
plt.ylabel('Anzahl der Einkäufe')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 11. Korrelation zwischen Einkaufswert und Anzahl der Einkäufe

# Jointplot für die Korrelation zwischen Einkaufswert und Anzahl der Einkäufe
plt.figure(figsize=(8, 6))

# Jointplot erstellen
sns.jointplot(x='Purchase_VALUE', y='N_Purchases', data=data, kind='reg', color='purple')

plt.suptitle('Korrelation zwischen Einkaufswert und Anzahl der Einkäufe', y=1.02)
plt.tight_layout()
plt.show()

# Erstellung des Jointplots
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
sns.scatterplot(x='Purchase_VALUE', y='N_Purchases', data=data, color='blue', alpha=0.6)
sns.regplot(x='Purchase_VALUE', y='N_Purchases', data=data, scatter=False, color='purple')

# Hinzufügen von Titeln und Beschriftungen
plt.title('Korrelation zwischen Einkaufswert und Anzahl der Einkäufe', fontsize=14)
plt.xlabel('Einkaufswert', fontsize=12)
plt.ylabel('Anzahl der Einkäufe', fontsize=12)
plt.tight_layout()
plt.show()

# Histogramm für die Spalte 'Purchase_VALUE'
plt.figure(figsize=(8, 6))
sns.histplot(data['Purchase_VALUE'], bins=30, kde=True, color='skyblue')
plt.title('Verteilung des Einkaufswerts', fontsize=14)
plt.xlabel('Einkaufswert', fontsize=12)
plt.ylabel('Anzahl', fontsize=12)
plt.tight_layout()
plt.show()

# 12. Die Verteilung von 'Revenue_Total

# Statistische Kennzahlen berechnen
min_value = data['Revenue_Total'].min()
max_value = data['Revenue_Total'].max()
mean_value = data['Revenue_Total'].mean()

# Setzen des Stils für die Diagramme
sns.set(style="whitegrid")

# Erstellung von Subplots für Histogramm und Boxplot
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Histogramm für die Verteilung von 'Revenue_Total'
sns.histplot(data['Revenue_Total'], kde=True, color='skyblue', ax=axs[0])
axs[0].set_title('Verteilung von Revenue_Total (Histogramm)', fontsize=14)
axs[0].set_xlabel('Revenue_Total', fontsize=12)
axs[0].set_ylabel('Anzahl', fontsize=12)

# Text für statistische Kennzahlen im Histogramm
axs[0].text(0.65, 0.85, f'Min: {min_value:.2f}\nMax: {max_value:.2f}\nMean: {mean_value:.2f}',
             transform=axs[0].transAxes, fontsize=10, verticalalignment='top')

# Boxplot für die Verteilung von 'Revenue_Total'
sns.boxplot(x=data['Revenue_Total'], color='lightgreen', ax=axs[1])
axs[1].set_title('Verteilung von Revenue_Total (Boxplot)', fontsize=14)
axs[1].set_xlabel('Revenue_Total', fontsize=12)

# Text für statistische Kennzahlen im Boxplot
axs[1].text(0.65, 0.85, f'Min: {min_value:.2f}\nMax: {max_value:.2f}\nMean: {mean_value:.2f}',
             transform=axs[1].transAxes, fontsize=10, verticalalignment='top')

plt.tight_layout()
plt.show()

# Ausgabe der statistischen Kennzahlen
print(f"Min: {min_value:.2f}")
print(f"Max: {max_value:.2f}")
print(f"Mean: {mean_value:.2f}")


# 13. Korrelationsmatrix zwischen Zeit, Kaufbetrag und Ausgaben

# Berechnung der Korrelationsmatrix für Zeit, Kaufbetrag und Ausgaben
correlation_matrix_1 = data[['Time_Spent', 'Revenue_Total', 'Purchase_VALUE']].corr()

plt.figure(figsize=(16, 6))

# Erste Heatmap: Korrelationsmatrix zwischen Zeit, Kaufbetrag und Ausgaben
plt.subplot(1, 2, 1)
sns.heatmap(correlation_matrix_1, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"fontsize": 12})
plt.title('Korrelationsmatrix zwischen Zeit, Kaufbetrag und Ausgaben', fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Auswahl der relevanten Spalten für die zweite Korrelationsmatrix
selected_columns = ['N_Purchases', 'Purchase_VALUE', 'Time_Spent']

# Berechnung der Korrelationsmatrix für N_Purchases, Purchase_VALUE und Time_Spent
correlation_matrix_2 = data[selected_columns].corr()

# Zweite Heatmap: Korrelations-Heatmap zwischen N_Purchases, Purchase_VALUE und Time_Spent
plt.subplot(1, 2, 2)
sns.heatmap(correlation_matrix_2, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korrelations-Heatmap zwischen N_Purchases, Purchase_VALUE und Time_Spent', fontsize=14)

plt.tight_layout()
plt.show()
# Berechnung der Korrelationsmatrix für Zeit, Kaufbetrag und Ausgaben
correlation_matrix_1 = data[['Time_Spent', 'Revenue_Total', 'Purchase_VALUE']].corr()

# Ausgabe der Korrelationsmatrix für Zeit, Kaufbetrag und Ausgaben
print("Korrelationsmatrix zwischen Zeit, Kaufbetrag und Ausgaben:")
print(correlation_matrix_1)

# Auswahl der relevanten Spalten für die zweite Korrelationsmatrix
selected_columns = ['N_Purchases', 'Purchase_VALUE', 'Time_Spent']

# Berechnung der Korrelationsmatrix für N_Purchases, Purchase_VALUE und Time_Spent
correlation_matrix_2 = data[selected_columns].corr()

# Ausgabe der Korrelationsmatrix für N_Purchases, Purchase_VALUE und Time_Spent
print("\nKorrelationsmatrix zwischen N_Purchases, Purchase_VALUE und Time_Spent:")
print(correlation_matrix_2)


# 14. import pandas as pd

# Berechnung des Gewinns ('Profit') basierend auf 'Revenue_Total' und 'Purchase_VALUE'
data['Profit'] = data['Revenue_Total'] - data['Purchase_VALUE']

# Erstellen eines Histogramms für den berechneten Gewinn ('Profit')
plt.figure(figsize=(8, 6))
plt.hist(data['Profit'], bins=30, color='skyblue', alpha=0.7)
plt.title('Histogramm des Gewinns')
plt.xlabel('Gewinn')
plt.ylabel('Anzahl')
plt.grid(True)
plt.tight_layout()
plt.show()

# Erstellen eines Boxplots für den berechneten Gewinn ('Profit')
plt.figure(figsize=(8, 6))
plt.boxplot(data['Profit'], vert=False)
plt.title('Boxplot des Gewinns')
plt.xlabel('Gewinn')
plt.yticks([])
plt.grid(True)
plt.tight_layout()
plt.show()

# Berechnung von Min, Max und Mean des Profits
min_profit = data['Profit'].min()
max_profit = data['Profit'].max()
mean_profit = data['Profit'].mean()

# Ausgabe der statistischen Informationen
print(f"Min. Gewinn: {min_profit}")
print(f"Max. Gewinn: {max_profit}")
print(f"Durchschnittlicher Gewinn: {mean_profit:.2f}")

# 14.1 Null-Gewinn:

min_profit = data['Profit'].min()
max_profit = data['Profit'].max()
mean_profit = data['Profit'].mean()

# Filtern der Daten für Null-Gewinne
zero_profit_data = data[data['Profit'] == 0]

# Anzahl der Fälle mit Null-Gewinnen
num_zero_profits = len(zero_profit_data)

# Histogramm für den Gewinn erstellen
plt.figure(figsize=(8, 6))
plt.hist(data['Profit'], bins=30, color='skyblue', alpha=0.7)
plt.title('Histogramm des Gewinns')
plt.xlabel('Gewinn')
plt.ylabel('Anzahl')
plt.grid(True)
plt.axvline(x=0, color='red', linestyle='--', label=f'Null-Gewinn: {num_zero_profits}')
plt.legend()

# Text für Min, Max und Mean hinzufügen
plt.text(min_profit, 200, f"Min: {min_profit}", ha='center', va='center')
plt.text(max_profit, 200, f"Max: {max_profit}", ha='center', va='center')
plt.text(mean_profit, 200, f"Mean: {mean_profit:.2f}", ha='center', va='center')

plt.tight_layout()
plt.show()


#15. Zahlungsmethoden nach Geschlecht
onlineShop['Pay_Method'] = onlineShop['Pay_Method'].replace({
    0: 'Karte',
    1: 'PayPal',
    2: 'Digitale Geldbörsen',
    3: 'Andere'
})
pay_method_gender_pivot = data.pivot_table(index='Gender', columns='Pay_Method', values='Age', aggfunc='count')

# Erstellung der Heatmap für Zahlungsmethoden nach Geschlecht
plt.figure(figsize=(8, 6))
sns.heatmap(pay_method_gender_pivot, annot=True, cmap='YlGnBu')
plt.title('Zahlungsmethoden nach Geschlecht (Heatmap)')
plt.xlabel('Zahlungsmethoden')
plt.ylabel('Geschlecht')
plt.tight_layout()
plt.show()

#15.1 Zahlungsmethoden nach Altersgruppe:

# Erstellen einer neuen Spalte "Age_Group" basierend auf Altersgruppen
bins = [0, 18, 30, 50, 100]  # Altersgruppen definieren
labels = ['<18', '18-30', '31-50', '50+']  # Altersgruppenbezeichnungen
data['Age_Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Pivot-Tabelle für Zahlungsmethoden nach Altersgruppe
onlineShop['Pay_Method'] = onlineShop['Pay_Method'].replace({
    0: 'Karte',
    1: 'PayPal',
    2: 'Digitale Geldbörsen',
    3: 'Andere'
})
pay_method_age_group_pivot = data.pivot_table(index='Age_Group', columns='Pay_Method', values='Gender', aggfunc='count')

# Erstellung der Heatmap für Zahlungsmethoden nach Altersgruppe
plt.figure(figsize=(10, 8))
sns.heatmap(pay_method_age_group_pivot, annot=True, cmap='YlGnBu')
plt.title('Zahlungsmethoden nach Altersgruppe (Heatmap)')
plt.xlabel('Zahlungsmethoden')
plt.ylabel('Altersgruppe')
plt.tight_layout()
plt.show()

```

```python
# Resampling data by quarter and month to calculate the total revenue for each period
data.set_index('Purchase_DATE', inplace=True)
revenue_over_time_quarterly = data['Revenue_Total'].resample('Q').sum()
revenue_over_time_monthly = data['Revenue_Total'].resample('M').sum()

# Calculating min, max, and mean for quarterly revenue
quarterly_min = revenue_over_time_quarterly.min()
quarterly_max = revenue_over_time_quarterly.max()
quarterly_mean = revenue_over_time_quarterly.mean()

# Calculating min, max, and mean for monthly revenue
monthly_min = revenue_over_time_monthly.min()
monthly_max = revenue_over_time_monthly.max()
monthly_mean = revenue_over_time_monthly.mean()

# Creating subplots with shared x-axis
fig, axs = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Plotting quarterly revenue
axs[0].plot(revenue_over_time_quarterly.index, revenue_over_time_quarterly.values, marker='o', linestyle='-')
axs[0].set_ylabel('Total Revenue (Quarterly)')
axs[0].set_title('Total Revenue Over Time (Quarterly)\nMin: {:.2f}, Max: {:.2f}, Mean: {:.2f}'.format(quarterly_min, quarterly_max, quarterly_mean))
axs[0].tick_params(axis='x', rotation=45)

# Plotting monthly revenue
axs[1].plot(revenue_over_time_monthly.index, revenue_over_time_monthly.values, marker='o', linestyle='-')
axs[1].set_xlabel('Purchase Date')
axs[1].set_ylabel('Total Revenue (Monthly)')
axs[1].set_title('Total Revenue Over Time (Monthly)\nMin: {:.2f}, Max: {:.2f}, Mean: {:.2f}'.format(monthly_min, monthly_max, monthly_mean))
axs[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
# Printing statistics for quarterly revenue
print("Quarterly Revenue Statistics:")
print("Min:", quarterly_min)
print("Max:", quarterly_max)
print("Mean:", quarterly_mean)

# Printing statistics for monthly revenue
print("\nMonthly Revenue Statistics:")
print("Min:", monthly_min)
print("Max:", monthly_max)
print("Mean:", monthly_mean)

```
